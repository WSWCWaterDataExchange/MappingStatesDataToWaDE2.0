```python
#Needed Libararies
import os
import numpy as np
import pandas as pd
from datetime import datetime
import geopandas as gpd # the library that lets us read in shapefiles
pd.set_option('display.max_columns', 999)  # How to display all columns of a Pandas DataFrame in Jupyter Notebook
```

```python
#Setting work directory, reading inputs, creating dataframe

#CSV input file
workingDir = "C:/Users/rjame/Documents/WSWC Documents/MappingStatesDataToWaDE2.0/USBR/Colorado River Basin Natural Flow and Salt Data/Orange Data Source Sheets/AggregatedAmounts"
os.chdir(workingDir)
fileInput = "RawInputData/perfect example OT_A.xlsx"
df = pd.read_excel(fileInput, sheet_name='HUC Level')

#County Shapefile input
shfileCounty = gpd.read_file('C:/Users/rjame/Documents/WSWC Documents/MappingStatesDataToWaDE2.0/USBR/Colorado River Basin Natural Flow and Salt Data/Orange Data Source Sheets/AggregatedAmounts/RawInputData/COUNTY_Shapefile/cb_2018_us_county_500k.shp')
dfshapeCounty = pd.DataFrame(shfileCounty)

#HUC8 Shapefile input
shfileHUC8 = gpd.read_file('C:/Users/rjame/Documents/WSWC Documents/MappingStatesDataToWaDE2.0/USBR/Colorado River Basin Natural Flow and Salt Data/Orange Data Source Sheets/AggregatedAmounts/RawInputData/HUC8_Shapefile/HUC8_US.shp')
dfshapeHUC8 = pd.DataFrame(shfileHUC8)
```

```python
#Approach:
#Make a temp df to store in values from a list
#Append temp df to output df

columnsList = ["STATE", "HUC8", "COUNTY", "RESERVOIRNAME", "Amount", "ReportYearCV", "BeneficialUseCV"]
df2 = pd.DataFrame(columns=columnsList)
dftemp = pd.DataFrame(columns=columnsList)

# for row in df.index:
for row in df.index:
    #Get Values
    State = str(df.iloc[row,0])
    Huc8 = str(df.iloc[row,1])
    County = str(df.iloc[row,2])
    Reservoir = str(df.iloc[row,3])
    BenUse = df.iloc[row, 4]
    Amounts = df.iloc[row].drop(["State", "HUC8", "County", "ReservoirName", "BenUse"]).values.tolist()
    ReportYears = df.iloc[0].drop(["State", "HUC8", "County", "ReservoirName", "BenUse"]).index.values.tolist()
    
    #Put Values into temp Dataframe
    dftemp['STATE'] = State
    dftemp['HUC8'] = Huc8
    dftemp['COUNTY'] = County
    dftemp['RESERVOIRNAME'] = Reservoir
    dftemp['BeneficialUseCV'] = BenUse
    dftemp['Amount'] = Amounts
    dftemp['ReportYearCV'] = ReportYears
  
    #Append to dfout
    df2 = df2.append(dftemp)

df2 = df2.reset_index()
df2
```

```python
df2 = df2[df2['STATE'].notna()]
df2
```

```python
#Fixing Datatypes
df2['Amount'] = df2['Amount'].astype(float) 
df2['ReportYearCV'] = df2['ReportYearCV'].astype(int)
df2.dtypes
```

### Seperate out into County vs HUC8

```python
#County
dfcounty = df2
dfcounty = dfcounty.drop(['HUC8'], axis=1)
dfcounty = dfcounty[dfcounty['COUNTY'] != 'nan']
dfcounty['COUNTY'] = dfcounty['COUNTY'].str.title()
dfcounty['TheJoin'] = dfcounty['STATE'] + dfcounty['COUNTY']
dfcounty = dfcounty.reset_index(drop=True)
dfcounty
```

```python
dfcounty.to_csv('RawInputData/temp_county.csv', index=False)
```

```python
#transfer WKT gemetry from dfshapeCounty


def retrieveGeometry(colrowValue, dfshapeCounty):
    if (colrowValue == '') or (pd.isnull(colrowValue)):
        outList = ''
    else:
        ml = dfshapeCounty.loc[(dfshapeCounty['NAME'] == colrowValue), 'geometry']
        if not(ml.empty):  # check if the series is empty
            outList = ml.iloc[0]
        else:
            outList = ''
    return outList

dfcounty['Geometry'] = dfcounty.apply(lambda row: retrieveGeometry(row['COUNTY'], dfshapeCounty), axis=1)
dfcounty
```

```python
#HUC8
dfHUC8 = df2
dfHUC8 = dfHUC8.drop(['COUNTY'], axis=1)
dfHUC8 = dfHUC8.dropna(axis=0, subset=['HUC8'])
dfHUC8 = dfHUC8.reset_index(drop=True)
dfHUC8
```

```python
#transfer WKT gemetry from dfshapeHUC8
def retrieveGeometry(colrowValue, dfshapeHUC8):
    if (colrowValue == '') or (pd.isnull(colrowValue)):
        outList = ''
    else:
        ml = dfshapeHUC8.loc[(dfshapeHUC8['HUC8'] == colrowValue), 'geometry']
        if not(ml.empty):  # check if the series is empty
            outList = ml.iloc[0]
        else:
            outList = ''
    return outList

dfHUC8['Geometry'] = dfHUC8.apply(lambda row: retrieveGeometry(row['HUC8'], dfshapeHUC8), axis=1)
dfHUC8
```

### Merge and the Outpt

```python
columnsList = ['STATE', 'ReportingUnit', 'RESERVOIRNAME', 'Amount', 'ReportYearCV', 'BeneficialUseCV', 'Geometry']
dfout = pd.DataFrame(columns=columnsList)

STATEframes = [dfcounty['STATE'], dfHUC8['STATE']]
dfout['STATE'] = pd.concat(STATEframes)

ReportingUnitframes = [dfcounty['COUNTY'], dfHUC8['HUC8']]
dfout['ReportingUnit'] = pd.concat(ReportingUnitframes)

RESERVOIRNAMEframes = [dfcounty['RESERVOIRNAME'], dfHUC8['RESERVOIRNAME']]
dfout['RESERVOIRNAME'] = pd.concat(RESERVOIRNAMEframes)

Amountframes = [dfcounty['Amount'], dfHUC8['Amount']]
dfout['Amount'] = pd.concat(Amountframes)

ReportYearCVframes = [dfcounty['ReportYearCV'], dfHUC8['ReportYearCV']]
dfout['ReportYearCV'] = pd.concat(ReportYearCVframes)

BeneficialUseCVframes = [dfcounty['BeneficialUseCV'], dfHUC8['BeneficialUseCV']]
dfout['BeneficialUseCV'] = pd.concat(BeneficialUseCVframes)

Geometryframes = [dfcounty['Geometry'], dfHUC8['Geometry']]
dfout['Geometry'] = pd.concat(Geometryframes)


dfout = dfout.reset_index(drop=True)
dfout
```
