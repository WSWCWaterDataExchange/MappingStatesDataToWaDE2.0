```python
#Input files - Water Usage
wu_input = "water_data_usage_input.csv"
df = pd.read_csv(wu_input)
print(len(df))
df.head(3)
```

```python
#Input files - Jim Vohden Notes
VJ_input = "Jim Vohden Notes_input.csv"
dfJV = pd.read_csv(VJ_input)
print(len(dfJV))
dfJV.head(3)
```

```python
#Create new temp dataframe with time series data removed.  Only interested in water right info at this time.
dfwu = df.drop(['RPT_YEAR', 'RPT_MONTH', 'WATER_SOURCE_USAGE_ID', 'ENTERED_TIME', 'QUANTITY', 'QTY_UNIT', 'DAILY_PEAK_QTY'], axis=1)
dfwu = dfwu.drop_duplicates()
print(len(dfwu))
dfwu.head(3)
```

```python
#Update datatype of Priority Date to fit WaDE 2.0 structure
dfJV['Priority Date'] = pd.to_datetime(dfJV['Priority Date'])
dfJV['Priority Date'] = pd.to_datetime(dfJV["Priority Date"].dt.strftime('%m/%d/%Y'))
dfJV.head(3)
```

```python
#Left join dataframes
dfout = pd.merge(dfwu, dfJV, left_on='FILE_NUMBER', right_on='FILE_NUMBER', how='left')
dfout
```

```python
# # Checking Input
# siteList = dfout['INTAKE_LOCATION'].tolist()
# siteList
```

```python
def retrieveLocation(colrowValue):
    if colrowValue == '' or pd.isnull(colrowValue):
        outStringLat =  ""
        outStringLong = ""
    else:
        outStringLat =  ""
        outStringLong = ""
        s = str(colrowValue)
        s = s.replace('POINT (', '')
        s = s.replace(')', '')
        x = s.split(' ')
        try:
            outList = []
            for i in range(len(x)):
                y = x[i].strip()
                outList.append(y)
            outStringLat = str(outList[0])
            outStringLong = str(outList[1])
        except:
            outStringLat =  ""
            outStringLong = ""
    return pd.Series([outStringLat,  outStringLong])

dfout[["wswc_long", "wswc_lat"]] = dfout.apply(lambda row: retrieveLocation(row['INTAKE_LOCATION']), axis=1)
dfout.head(3)
```

```python
#converting *TOTAL_PERMIT_QTY* from gallons to AF.

def convertQTY(colrowValue):
    if colrowValue == '' or pd.isnull(colrowValue):
        outString = ''
    else:
        try:
            outString = colrowValue * 3.06889e-6
        except:
            outString = ''
    return outString

dfout['wswc_QTY'] = dfout.apply(lambda row: convertQTY(row['TOTAL_PERMIT_QTY']), axis=1)
dfout.head(3)
```

## WaDE Custom Elements (due to missing sate info)

```python
# Creating WaDE Custom site native ID for easy site identificaiion
# ----------------------------------------------------------------------------------------------------

# Create temp SiteNativeID dataframe of unique site.
def assignSiteUUID(colrowValue):
    string1 = str(colrowValue)
    outstring = "WaDEAK_S" + string1
    return outstring

dfSiteNativeID = pd.DataFrame()
dfSiteNativeID['wswc_Latitude'] = dfout['wswc_lat']
dfSiteNativeID['wswc_Longitude'] = dfout['wswc_long']
dfSiteNativeID['wswc_SiteName'] = dfout['INTAKE_NAME']
dfSiteNativeID = dfSiteNativeID.drop_duplicates()

dftemp = pd.DataFrame(index=dfSiteNativeID.index)
dftemp["Count"] = range(1, len(dftemp.index) + 1)
dfSiteNativeID['wswc_SiteNativeID'] = dftemp.apply(lambda row: assignSiteUUID(row['Count']), axis=1)

# ----------------------------------------------------------------------------------------------------

# Retreive WaDE Custom site native ID
def retrieveSiteNativeID(A, B, C):
    if (A == '' and B == '') or (pd.isnull(A) and pd.isnull(B)):
        outList = ''
    else:
        ml = dfSiteNativeID.loc[(dfSiteNativeID['wswc_Latitude'] == A) & 
                                (dfSiteNativeID['wswc_Longitude'] == B) &
                                (dfSiteNativeID['wswc_SiteName'] == C), 'wswc_SiteNativeID']
        if not (ml.empty):  # check if the series is empty
            outList = ml.iloc[0]
        else:
            outList = ''
    return outList

dfout['wswc_SiteNativeID'] = dfout.apply(lambda row: retrieveSiteNativeID( row['wswc_lat'], row['wswc_long'], row['INTAKE_NAME']), axis=1)
dfout.head(3)
```

```python
#Remove NA and NaN values
dfout = dfout.fillna('')
dfout = dfout.replace('NaN', '')
dfout.head(3)
```

```python
#Exporting to Finished File
dfout.to_csv('P_AlaskaMaster.csv', index=False)  # The output
```
