## Overlay Area Data #1

```python
# Input File
inputFile = "RawInputData/shapefiles/Basins_State_Engineer_Admin_Boundaries.zip"
dfin1 = gpd.read_file(inputFile).replace(np.nan, "")
dfin1['geometry'] = dfin1['geometry'].to_crs(epsg=4326) # Realign Geometry Projection

# WaDE UUID tracker for data assessment
if 'WaDEUUID' not in dfin1:
    dfin1['WaDEUUID'] = "ov" + dfin1.index.astype(str)
    dfin1.to_csv('RawInputData/Basins_State_Engineer_Admin_Boundaries.zip', compression=dict(method='zip', archive_name='Basins_State_Engineer_Admin_Boundaries.csv'), index=False)

print(len(dfin1))
dfin1.head(1)
```

```python
# create output for Regulatory Area #1 dataframe
df = pd.DataFrame()

# Data Assessment UUID
df['WaDEUUID'] = dfin1['WaDEUUID']

# Date Info
df['in_Date'] = "02/03/2025"
df['in_Year'] = "2025"

# Organization Info
df['in_OrganizationUUID'] = "NVov_O1"

# ReportingUnit Info
df['in_EPSGCodeCV'] = 4326
df['in_ReportingUnitName'] = dfin1['BasinName']
df['in_ReportingUnitNativeID'] =  "nv" + dfin1['BasinID']
df['in_ReportingUnitProductVersion'] = ""
df['in_ReportingUnitTypeCV'] = "Groundwater Basin Designations"
df['in_ReportingUnitUpdateDate'] = ""
df['in_StateCV'] = "NV"
df['in_Geometry'] = dfin1['geometry']

# RegulatoryOverlay Info
df['in_OversightAgency'] = "Office of the Nevada State Engineer"
df['in_RegulatoryDescription'] = "" # will fill in below
df['in_RegulatoryName'] = dfin1['BasinName']
df['in_RegulatoryOverlayNativeID'] = "nv" + dfin1['BasinID']
df['in_RegulatoryStatusCV'] = dfin1['Designatio']
df['in_RegulatoryStatute'] = ""
df['in_RegulatoryStatuteLink'] ="https://www.leg.state.nv.us/NRS/NRS-534.html#NRS534Sec037"
df['in_StatutoryEffectiveDate'] = "07/01/1981"
df['in_StatutoryEndDate'] = ""
df['in_RegulatoryOverlayTypeCV'] = "Groundwater Basin Designations"
df['in_WaterSourceTypeCV'] = "Groundwater"

df = df.drop_duplicates().reset_index(drop=True)
outdf1 = df.copy()
print(len(outdf1))
outdf1.head()
```

## Clean Data / data types

```python
outdf['in_RegulatoryStatusCV'].value_counts()
```

```python

def UpdateRegulatoryStatusCVFunc(valA):
    valA = valA.strip()
    if valA == "" or pd.isnull(valA):
        outString = "Not designated."
    else:
        outString = valA
    return outString

outdf['in_RegulatoryStatusCV'] = outdf.apply(lambda row: UpdateRegulatoryStatusCVFunc(row['in_RegulatoryStatusCV']), axis=1)
outdf['in_RegulatoryStatusCV'].value_counts()
```

```python
# Include RegulatoryDescription

UnknownDescription = {
"DESIGNATED" : "State Engineer's order(s) do not define any administrative controls.",
"DESIGNATED (ID)" : "State Engineer's order(s) state that irrigation is not a preferred use in these basins and applications for new irrigation appropriations will be denied.",
"DESIGNATED (PU)" : "State Engineer's order(s) list certain types of uses as preferred in these basins, and quantity restrictions may be placed on these preferred uses.",
"DESIGNATED (PUID)" : "State Engineer's order(s) list certain types of uses as preferred in these basins. Quantity restrictions may be placed on these preferred uses. State Engineer's order(s) also state that irrigation is not a preferred use in these basins and applications for new irrigation appropriations will be denied. Other uses may also be listed as denied.",
"PARTLY DESIGNATED" : "A portion of the groundwater basin has been officially designated by the State Engineer for special management due to concerns such as overuse or declining water levels. In these designated areas, stricter regulations and oversight are applied to groundwater extraction to ensure sustainable use.",
"PARTLY DESIG.(PU)" : "A portion of the basin is designated specifically concerning public utilities, where groundwater management practices are tailored to address the needs and impacts of public utility operations, ensuring that water extraction by utilities is sustainable and does not adversely affect the basin's overall health."
}

def assignUnknownDescription(valA):
    valA = valA.strip()
    if valA == "Not designated." or pd.isnull(valA):
        outString = "No description, not designated."
    else:
        outString = UnknownDescription[valA]
    return outString

outdf['in_RegulatoryDescription'] = outdf.apply(lambda row: assignUnknownDescription(row['in_RegulatoryStatusCV']), axis=1)
outdf['in_RegulatoryDescription'].value_counts()
```

```python
# Ensure Empty String / remove string value of "nan"

def ensureEmptyString(val):
    val = str(val).strip()
    if val == "" or val == " " or val == "nan" or pd.isnull(val):
        outString = ""
    else:
        outString = val
    return outString
```

```python
outdf['in_ReportingUnitName'] = outdf.apply(lambda row: ensureEmptyString(row['in_ReportingUnitName']), axis=1)
outdf['in_ReportingUnitName'].unique()
```

```python
outdf['in_RegulatoryDescription'] = outdf.apply(lambda row: ensureEmptyString(row['in_RegulatoryDescription']), axis=1)
outdf['in_RegulatoryDescription'].unique()
```

```python
outdf['in_RegulatoryName'] = outdf.apply(lambda row: ensureEmptyString(row['in_RegulatoryName']), axis=1)
outdf['in_RegulatoryName'].unique()
```

```python
#Update datatype of StatutoryEffectiveDate to fit WaDE 2.0 structure
outdf['in_StatutoryEffectiveDate'] = pd.to_datetime(outdf['in_StatutoryEffectiveDate'], errors = 'coerce')
outdf['in_StatutoryEffectiveDate'] = pd.to_datetime(outdf['in_StatutoryEffectiveDate'].dt.strftime('%m/%d/%Y'))
outdf['in_StatutoryEffectiveDate'].unique()
```
