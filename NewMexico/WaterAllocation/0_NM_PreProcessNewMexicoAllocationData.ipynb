```python
#Chaning datatype of date columns
def formatDateString(inString):
    #print(inString)
    try:
        if inString == '' or pd.isnull(inString):
            valndf = ''
        else:
            valD = datetime.strptime(inString, '%Y-%m-%dT00:00:00.000Z')
            #print(valD)
            valnDd = valD.date()
            #print(valnDd)
            valndf = valnDd.strftime('%m/%d/%Y')
            #print('date:', valndf)
    except:
        valndf = ''

    return valndf

df = df.assign(AllocationPriorityDate='')
df['AllocationPriorityDate'] = df.apply(lambda row: formatDateString(row['finish_dat']), axis=1)

df
```

```python
#creating Owner

def assignownerName(colrowValue1, colrowValue2):
    if colrowValue1 == '' or pd.isnull(colrowValue1):
        outList1 = ''
    else:
        outList1 = colrowValue1.strip()  # remove whitespace chars
    if colrowValue2 == '' or pd.isnull(colrowValue2):
        outList2 = ''
    else:
        outList2 = colrowValue2.strip()  # remove whitespace chars

    if outList1 == '' and outList2 == '':
        outList = ''
    elif outList1 == '':
        outList = outList2
    elif outList2 == '':
        outList = outList1
    else:
        outList = ",".join(map(str, [colrowValue1, colrowValue2]))
    return outList


df['AllocationOwner'] = df.apply(lambda row: assignownerName(row['own_lname'], row['own_fname']), axis=1)
df
```

```python
print("Project to longitude/ latitude ")

df = df.assign(Longitude='')
df = df.assign(Latitude='')

utmProj = Proj(init="epsg:26913", proj="utm", zone=13)

# drop cells with no x or y coordinate
df = df.replace(np.nan, '') 
dropIndex = df.loc[(df['easting'] == '') | (df['northing'] == '')].index
if len(dropIndex) > 0:
    df = df.drop(dropIndex)
    df = df.reset_index(drop=True)

lonList = []
latList = []

#for ix in range(len(df100.index)):
for ix in range(len(df.index)):
    #print(ix)
    x1 = df.loc[ix, 'easting']
    y1 = df.loc[ix, 'northing']
    try:
        lon, lat = utmProj(float(x1), float(y1), inverse=True)
        lonList.append(lon)
        latList.append(lat)
    except:
        lonList.append('')
        latList.append('')

df['Longitude'] = lonList
df['Latitude'] = latList

df
```

```python
print ("SiteName...")

# For creating WaterSourceNativeID
def assignSiteName(colrowValue):
    colrowValue = str(colrowValue)
    colrowValue = colrowValue.strip()
    if colrowValue == '' or colrowValue == ' ' or pd.isnull(colrowValue):
        outList = 'Unspecified'
    else:
        strvalue = str(colrowValue)
        outList = strvalue.strip()
    return outList

df['ditch_name'] = df.apply(lambda row: assignSiteName(row['ditch_name']), axis=1)

#df["ditch_name"] = df['ditch_name'].astype(str)

df = df.replace(np.nan, '')
 # If site name doesn't exist use Not provided
df["SiteName"] = df.apply(lambda row: row["ditch_name"] 
                                                   if str(row["ditch_name"]).strip() != '' 
                                                   else "Unspecified" , axis=1)

df 
```

```python
print ("SiteTypeCV...")

df.assign(GroundWaterSiteTypeCV='')
df.assign(SurfaceWaterSiteTypeCV='')

df["GroundWaterSiteTypeCV"] = df.apply(lambda row: 'Ground Water' 
                                             if str(row["grnd_wtr_s"]).strip() != '' 
                                             else '', axis=1)
df["SurfaceWaterSiteTypeCV"] = df.apply(lambda row: 'Surface Water' 
                                              if int(row["surface_co"]) > 0 
                                              else '', axis=1)


df.assign(SiteTypeCV='')

siteTypeList = []
for index, row in df.iterrows():
    #print(index)
    groundWatetrStr = row['GroundWaterSiteTypeCV']
    surfaceWatetrStr = row['SurfaceWaterSiteTypeCV']
    if (groundWatetrStr != '') & (surfaceWatetrStr != ''):
        siteTypeList.append('Surfacewater_Groundwater')
    elif groundWatetrStr != '':
        siteTypeList.append('Ground Water')
    elif surfaceWatetrStr != '':
        siteTypeList.append('Surface Water')
    else:
        siteTypeList.append("Unknown") #hard code "Unknown" for SiteTypeCV if it is missing

df['SiteTypeCV'] = siteTypeList
df
```

```python
print ("Water source type")

df.assign(GroundWaterSourceTypeCV='')
df.assign(SurfaceWaterSourceTypeCV='')

df["GroundWaterSourceTypeCV"] = df.apply(lambda row: 'Groundwater' 
                                             if str(row["grnd_wtr_s"]).strip() != '' 
                                             else '', axis=1)
df["SurfaceWaterSourceTypeCV"] = df.apply(lambda row: 'Surface Water' 
                                              if int(row["surface_co"]) > 0 
                                              else '', axis=1)



df.assign(WaterSourceTypeCV='')

waterSourceList = []
for index, row in df.iterrows():
    #print(index)
    groundWatetrStr = row['GroundWaterSourceTypeCV']
    surfaceWatetrStr = row['SurfaceWaterSourceTypeCV']
    if (groundWatetrStr != '') & (surfaceWatetrStr != ''):
        waterSourceList.append(",".join([groundWatetrStr, surfaceWatetrStr]))
    elif groundWatetrStr != '':
        waterSourceList.append(groundWatetrStr)
    elif surfaceWatetrStr != '':
        waterSourceList.append(surfaceWatetrStr)
    else:
        waterSourceList.append("Unknown")

df['WaterSourceTypeCV'] = waterSourceList


df
```
