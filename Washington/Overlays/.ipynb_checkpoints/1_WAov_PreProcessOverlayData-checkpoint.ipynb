## Overlay Area Data #1 - WDNR

```python
# Input File
inputFile = "RawInputData/WDNR/shapefiles/WatershedAdministrativeUnits_ForestPracticesRegulation.zip"
dfin1 = gpd.read_file(inputFile).replace(np.nan, "")
dfin1['geometry'] = dfin1['geometry'].to_crs(epsg=4326) # Realign Geometry Projection

# WaDE UUID tracker for data assessment
if 'WaDEUUID' not in dfin1:
    dfin1['WaDEUUID'] = "wa1" + dfin1.index.astype(str)
    dfin1.to_csv('RawInputData/WatershedAdministrativeUnits_ForestPracticesRegulation.zip', compression=dict(method='zip', archive_name='WatershedAdministrativeUnits_ForestPracticesRegulation.csv'), index=False)

print(len(dfin1))
dfin1.head()
```

```python
# create output for Regulatory Area #1 dataframe
outdf1 = pd.DataFrame()

# Data Assessment UUID
outdf1['WaDEUUID'] = dfin1['WaDEUUID']

# Date Info
outdf1['in_Date'] = "8/2/2023"
outdf1['in_Year'] = "2023"

# Organization Info
outdf1['in_OrganizationUUID'] = "WAov_O1"

# ReportingUnit Info
outdf1['in_EPSGCodeCV'] = "4326"
outdf1['in_ReportingUnitName'] = dfin1["WAU_LABEL_"]
outdf1['in_ReportingUnitNativeID'] = "wdnr" + dfin1["WAU_ID"].astype(int).astype(str).str.strip()
outdf1['in_ReportingUnitProductVersion'] = ""
outdf1['in_ReportingUnitTypeCV'] = "Watershed Administrative Units"
outdf1['in_ReportingUnitUpdateDate'] = "1/7/2021"
outdf1['in_StateCV'] = "WA"
outdf1['in_Geometry'] = dfin1['geometry']

# Overlay Info
outdf1['in_OversightAgency'] = "Washington Department of Natural Resources"
outdf1['in_OverlayDescription'] = "a structured approach to developing a forest practices plan for a WAU based on a biological and physical inventory"
outdf1['in_OverlayName'] = dfin1["WAU_LABEL_"]
outdf1['in_OverlayNativeID'] = "wdnr" + dfin1["WAU_ID"].astype(int).astype(str).str.strip()
outdf1['in_OverlayStatusCV'] = "Active"
outdf1['in_Statute'] = ""
outdf1['in_StatuteLink'] =""
outdf1['in_StatutoryEffectiveDate'] = "8/2/1992"
outdf1['in_StatutoryEndDate'] = ""
outdf1['in_OverlayTypeCV'] = "Watershed Administrative Units"
outdf1['in_WaterSourceTypeCV'] = "Surface Water"

outdf1 = outdf1.drop_duplicates().reset_index(drop=True)
print(len(outdf1))
outdf1.head()
```

## Overlay Area Data #2 - WDOE

```python
# Input File
inputFile = "RawInputData/WDOE/shapefiles/WRIA.zip"
dfin2 = gpd.read_file(inputFile).replace(np.nan, "")
dfin2['geometry'] = dfin2['geometry'].to_crs(epsg=4326) # Realign Geometry Projection

# WaDE UUID tracker for data assessment
if 'WaDEUUID' not in dfin2:
    dfin2['WaDEUUID'] = "wa2" + dfin2.index.astype(str)
    dfin2.to_csv('RawInputData/Water_Resource_Inventory_Areas_(WRIA).zip', compression=dict(method='zip', archive_name='Water_Resource_Inventory_Areas_(WRIA).csv'), index=False)

print(len(dfin2))
dfin2.head()
```

```python
# create output for Regulatory Area #1 dataframe
outdf2 = pd.DataFrame()

# Data Assessment UUID
outdf2['WaDEUUID'] = dfin2['WaDEUUID']

# Date Info
outdf2['in_Date'] = "8/2/2023"
outdf2['in_Year'] = "2023"

# Organization Info
outdf2['in_OrganizationUUID'] = "WAov_O2"

# ReportingUnit Info
outdf2['in_EPSGCodeCV'] = "4326"
outdf2['in_ReportingUnitName'] = dfin2["WRIA_NM"]
outdf2['in_ReportingUnitNativeID'] = "wdoe" + dfin2["OBJECTID"].astype(str)
outdf2['in_ReportingUnitProductVersion'] = ""
outdf2['in_ReportingUnitTypeCV'] = "Water Resource Inventory Areas"
outdf2['in_ReportingUnitUpdateDate'] = dfin2["last_edi_1"]
outdf2['in_StateCV'] = "WA"
outdf2['in_Geometry'] =  dfin2['geometry']

# Overlay Info
outdf2['in_OversightAgency'] = "Washington Department of Ecology"
outdf2['in_OverlayDescription'] = "Our innovative partnerships sustain healthy land, air, and water in harmony with a strong economy."
outdf2['in_OverlayName'] = dfin2["WRIA_NM"]
outdf2['in_OverlayNativeID'] = "wdoe" + dfin2["OBJECTID"].astype(str)
outdf2['in_OverlayStatusCV'] = "Active"
outdf2['in_Statute'] = ""
outdf2['in_StatuteLink'] =""
outdf2['in_StatutoryEffectiveDate'] = "1/1/1971"
outdf2['in_StatutoryEndDate'] = ""
outdf2['in_OverlayTypeCV'] = "Water Resource Inventory Area"
outdf2['in_WaterSourceTypeCV'] = "Surface Water"

outdf2 = outdf2.drop_duplicates().reset_index(drop=True)
print(len(outdf1))
outdf1.head()
```

## Concatenate DataFrames together

```python
# Concatenate dataframes
frames = [outdf1, outdf2] # list all out dataframes here
outdf = pd.concat(frames)
outdf = outdf.drop_duplicates().reset_index(drop=True).replace(np.nan, "")
print(len(outdf))
```

## Clean Data / data types

```python
# Ensure Empty String / remove string value of "nan"

def ensureEmptyString(val):
    val = str(val).strip()
    if val == "" or val == " " or val == "nan" or pd.isnull(val):
        outString = ""
    else:
        outString = val
    return outString
```

```python
outdf['in_ReportingUnitName'] = outdf.apply(lambda row: ensureEmptyString(row['in_ReportingUnitName']), axis=1)
outdf['in_ReportingUnitName'].unique()
```

```python
outdf['in_OverlayDescription'] = outdf.apply(lambda row: ensureEmptyString(row['in_OverlayDescription']), axis=1)
outdf['in_OverlayDescription'].unique()
```

```python
outdf['in_OverlayName'] = outdf.apply(lambda row: ensureEmptyString(row['in_OverlayName']), axis=1)
outdf['in_OverlayName'].unique()
```

```python
#Update datatype of StatutoryEffectiveDate to fit WaDE 2.0 structure
outdf['in_StatutoryEffectiveDate'] = pd.to_datetime(outdf['in_StatutoryEffectiveDate'], errors = 'coerce')
outdf['in_StatutoryEffectiveDate'] = pd.to_datetime(outdf['in_StatutoryEffectiveDate'].dt.strftime('%m/%d/%Y'))
outdf['in_StatutoryEffectiveDate'].unique()
```

# Geometry Data
- For attaching geometry to overlay reporting unit area info.

```python
# # Input File / or use same input as above

gdfin = outdf.copy()
gdfin = gpd.GeoDataFrame(gdfin, geometry=gdfin['in_Geometry'], crs="EPSG:4326") # covert to geodataframe
print(len(gdfin))
gdfin.head()
```

```python
# plot shape info to map
gdfin.plot()
```

```python
# create output for Regulatory Area #1 dataframe
df = pd.DataFrame()

columnsList = ['in_ReportingUnitNativeID', 'geometry']
goutdf = pd.DataFrame(columns=columnsList, index=gdfin.index)

goutdf['in_ReportingUnitNativeID'] =  gdfin["in_ReportingUnitNativeID"].astype(str)  #in_ReportingUnitNativeID needs to match source from above equivalent dataframe
goutdf['geometry'] = gdfin['in_Geometry']
goutdf = goutdf.drop_duplicates().reset_index(drop=True)

# drop geometery from outdf
outdf = outdf.drop(['in_Geometry'], axis=1)

print(len(goutdf))
goutdf.head()
```
