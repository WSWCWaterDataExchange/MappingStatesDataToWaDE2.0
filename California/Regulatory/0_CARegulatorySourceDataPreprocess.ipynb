# CA Regulatory Data

Preprocessing input data for a smoother upload experience of the state data to the WaDE 2.0 database.
Using geopandas to read in shp file, and coverting to WKT for ReportingUnit geometry.

Notes:
- WKT will be in it's own separate geometery dataframe.

```python
#Needed Libararies
import os
import numpy as np
import pandas as pd
from datetime import datetime
import geopandas as gpd # the library that lets us read in shapefiles
pd.set_option('display.max_columns', 999)  # How to display all columns of a Pandas DataFrame in Jupyter Notebook

#Setting work directory, reading inputs, creating dataframe
workingDir = "C:/Users/rjame/Documents/WSWC Documents/MappingStatesDataToWaDE2.0/California/Regulatory/RawInputData"
os.chdir(workingDir)
```

## Interstate Stream Compact Regions

```python
#CSV input file
fileInput = "sgma_2019_basin_prioritization_dashboard_data_input.csv"

df = pd.read_csv(fileInput)
print(len(df))
df.head(3)
```

```python
# For Creating OversightAgency

OversightAgencyDict = {
    "NRO" : "Northern Region Office",
    "SRO" : "Southern Region Office",
    "NCRO" : "North Central Region Office",
    "SCRO" : "South Central Region Office",}

def CreateOversightAgency(colrowValue):
    if colrowValue == "" or pd.isnull(colrowValue):
        outList = ""
    else:
        String1 = colrowValue
        try:
            outList = OversightAgencyDict[String1]
        except:
            outList = "Unspecified"
    return outList
df['in_OversightAgency'] = df.apply(lambda row: CreateOversightAgency(row['Region_Office']), axis=1)
df.head(3)
```

```python
# For Creating RegulatoryStatute

def CreateRegulatoryStatute(valA):
    if valA == "" or pd.isnull(valA):
        outString = "Unspecified"
    else:
        outString = str(valA).strip()
    return outString

df['in_RegulatoryStatute'] = df.apply(lambda row: CreateRegulatoryStatute(row['Priority']), axis=1)
df.head(3)
```

```python
df['in_RegulatoryStatute'].value_counts()
```

```python
# Interstate Stream Compact Regions Dataframe
# -------------------------------------------

# Create new dataframe
columnList = ['Basin_Subbasin_Number', 
              'Basin_Subbasin_Name', 
              'in_RegulatoryStatute', 
              "in_OversightAgency"]
df_out = pd.DataFrame(columns=columnList, index=df.index)

df_out['Basin_Subbasin_Number'] = df['Basin_Subbasin_Number']
df_out['Basin_Subbasin_Name'] = df['Basin_Subbasin_Name']
df_out['in_RegulatoryStatute'] = df['in_RegulatoryStatute']
df_out['in_OversightAgency'] = df['in_OversightAgency']


df_out = df_out.drop_duplicates().reset_index()

print(len(df_out))
df_out.head(3)
```

# Shapefile Data

```python
# Interstate Stream Compact Regions

ShapeFile = "ShapefileData/2019_SGMA_Basins.shp"
SGMAShape = gpd.read_file(ShapeFile)

print(len(SGMAShape))
SGMAShape.head()
```

```python
# Interstate Stream Compact Regions output shapefile dataframe
columnList = ['Basin_Subb', 
              'Basin_Su_1',
              'geometry']
df_SGMA_shp = pd.DataFrame(columns=columnList, index=SGMAShape.index)

df_SGMA_shp['Basin_Subb'] = SGMAShape['Basin_Subb']
df_SGMA_shp['Basin_Su_1'] = SGMAShape['Basin_Su_1'].str.title()
df_SGMA_shp['geometry'] = SGMAShape['geometry']

print(len(df_SGMA_shp))
df_SGMA_shp.head(3)
```
